```python
champ_window = Window.partitionBy('player_champ_name')
```

```python
champ_lanes = match_df_exploded.groupBy('player_champ_name', 'player_individual_position')\
    .count()\
    .withColumn('total_champ_game', sum(col('count')).over(champ_window))\
    .withColumn('percent_games', (col('count') / col('total_champ_game')) * 100)\
    .where(col('percent_games') > 35)\
    .orderBy('player_champ_name')

champ_lanes = champ_lanes.select(
    *[col(col_name).alias(f'{col_name}_filter') for col_name in champ_lanes.columns]
)

join_cond = (champ_lanes.player_champ_name_filter == match_df_exploded.player_champ_name) & (champ_lanes.player_individual_position_filter == match_df_exploded.player_individual_position)

match_df_exploded_champ_filtered = match_df_exploded.join(champ_lanes, join_cond, 'inner').select(match_df_exploded.columns)
```

```python
def compute_avg_percentile_from_champ_df(columns, df):
    agg_exprs = []
    for col_name in columns :
        agg_exprs.append(round(avg(col_name), 4).alias(f"avg_{col_name}"))
        for quartile, label in zip([0.25, 0.5, 0.75], ["Q1", "Q2", "Q3"]):
            agg_exprs.append(round(percentile(col_name, lit(quartile), lit(1)), 4).alias(f"{label}_{col_name}"))
    
    return df.groupBy("player_champ_name", "player_individual_position", "player_win")\
             .agg(*agg_exprs)\
             .orderBy(desc('player_individual_position'), desc('player_win'))


def compute_cols_per_minutes(columns, df):
    for col_name in columns:
        df = df.withColumn(col_name, round(col(col_name) / (col('game_duration') / 60), 3))

    return df


def display_columns_avg_percentile_from_champ_result(columns, df):
    champ_lanes = df.select(col("player_individual_position")).distinct().collect()
    
    for row_lane in champ_lanes:
        lane = row_lane['player_individual_position']
        print(f"Lane : \033[1m{lane}\033[0m")
        rows = df.where(col("player_individual_position") == lane).collect()
        for col_name in columns:
            print(f"\n\033[1m{col_name}\033[0m :\n")
            print(f"\033[1m{'Win':<46}\033[0m| \033[1m{'Loose':<46}\033[0m")
            print(f"{'Avg':<10}| {'Q1':<10}| {'Q2':<10}| {'Q3':<10}| {'Avg':<10}| {'Q1':<10}| {'Q2':<10}| {'Q3':<10}")
            print(f"{rows[0][f'avg_{col_name}']:<10.2f}| {rows[0][f'Q1_{col_name}']:<10.2f}| {rows[0][f'Q2_{col_name}']:<10.2f}| {rows[0][f'Q3_{col_name}']:<10.2f}| {rows[1][f'avg_{col_name}']:<10.2f}| {rows[1][f'Q1_{col_name}']:<10.2f}| {rows[1][f'Q2_{col_name}']:<10.2f}| {rows[1][f'Q3_{col_name}']:<10.2f}")
            print()
        print("\n____________________________________________________________________________________")


def split_multi_col_to_save_format(columns, df):
    computed_struct = [
        struct(
            lit(col_name).alias('column_stats'),
            col(f'avg_{col_name}').alias('AVG'),
            col(f'Q1_{col_name}').alias('Q1'),
            col(f'Q2_{col_name}').alias('Q2'),
            col(f'Q3_{col_name}').alias('Q3'),
        )
        for col_name in columns
    ]
    
    return df.withColumn('stats', explode(array(*computed_struct)))\
             .select(
                 col('player_champ_name'),
                 col('player_individual_position'),
                 col('player_win'),
                 col('stats.column_stats').alias('column_stats'),
                 col('stats.AVG').alias('AVG'),
                 col('stats.Q1').alias('Q1'),
                 col('stats.Q2').alias('Q2'),
                 col('stats.Q3').alias('Q3')
             )
```

### Pings distribution

```python
total_pings_df = match_df_exploded_champ_filtered.withColumn("player_pings_total", (col("player_pings_all_in") + 
                                              col("player_pings_assist") + 
                                              col("player_pings_command") + 
                                              col("player_pings_missing") + 
                                              col("player_pings_ennemy_vision") + 
                                              col("player_pings_hold") + 
                                              col("player_pings_get_back") + 
                                              col("player_pings_need_vision") + 
                                              col("player_pings_on_my_way") + 
                                              col("player_pings_push") + 
                                              col("player_pings_vision_cleared") + 
                                              col("player_pings_biasic")))

ping_cols = [
    "player_pings_total",
    'player_pings_all_in',
    'player_pings_assist',
    'player_pings_command',
    'player_pings_missing',
    'player_pings_ennemy_vision',
    'player_pings_hold',
    'player_pings_get_back',
    'player_pings_need_vision',
    'player_pings_on_my_way',
    'player_pings_push',
    'player_pings_vision_cleared',
    'player_pings_biasic'
]

total_pings_per_min_df = compute_cols_per_minutes(ping_cols, total_pings_df)
computed_pings_df = compute_avg_percentile_from_champ_df(ping_cols, total_pings_per_min_df)
pings_to_save_df = split_multi_col_to_save_format(ping_cols, computed_pings_df) 
```
