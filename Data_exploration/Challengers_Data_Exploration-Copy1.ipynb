```python
def compute_avg_percentile_from_df(columns, df):
    agg_exprs = []
    for col_name in columns :
        agg_exprs.append(round(avg(col_name), 4).alias(f"avg_{col_name}"))
        for quartile, label in zip([0.25, 0.5, 0.75], ["Q1", "Q2", "Q3"]):
            agg_exprs.append(round(percentile(col_name, lit(quartile), lit(1)), 4).alias(f"{label}_{col_name}"))
    
    return df.groupBy("player_individual_position", "player_win")\
             .agg(*agg_exprs)\
             .orderBy(desc('player_individual_position'), desc('player_win'))


def compute_cols_per_minutes(columns, df):
    for col_name in columns:
        df = df.withColumn(col_name, round(col(col_name) / (col('game_duration') / 60), 3))

    return df


def display_columns_avg_percentile_result(columns, df):
    lanes = [
        'TOP',
        'JUNGLE',
        'MIDDLE',
        'BOTTOM',
        'UTILITY'
    ]
    
    for lane in lanes:
        print(f"Lane : \033[1m{lane}\033[0m")
        rows = df.where(col("player_individual_position") == lane).collect()
        for col_name in columns:
            print(f"\n\033[1m{col_name}\033[0m :\n")
            print(f"\033[1m{'Win':<46}\033[0m| \033[1m{'Loose':<46}\033[0m")
            print(f"{'Avg':<10}| {'Q1':<10}| {'Q2':<10}| {'Q3':<10}| {'Avg':<10}| {'Q1':<10}| {'Q2':<10}| {'Q3':<10}")
            print(f"{rows[0][f'avg_{col_name}']:<10.2f}| {rows[0][f'Q1_{col_name}']:<10.2f}| {rows[0][f'Q2_{col_name}']:<10.2f}| {rows[0][f'Q3_{col_name}']:<10.2f}| {rows[1][f'avg_{col_name}']:<10.2f}| {rows[1][f'Q1_{col_name}']:<10.2f}| {rows[1][f'Q2_{col_name}']:<10.2f}| {rows[1][f'Q3_{col_name}']:<10.2f}")
            print()
        print("\n____________________________________________________________________________________")


def split_multi_col_to_save_format(columns, df):
    computed_struct = [
        struct(
            lit(col_name).alias('column_stats'),
            col(f'avg_{col_name}').alias('AVG'),
            col(f'Q1_{col_name}').alias('Q1'),
            col(f'Q2_{col_name}').alias('Q2'),
            col(f'Q3_{col_name}').alias('Q3'),
        )
        for col_name in columns
    ]
    
    return df.withColumn('stats', explode(array(*computed_struct)))\
             .select(
                 lit('GLOBAL').alias('player_champ_name'),
                 col('player_individual_position'),
                 col('player_win'),
                 col('stats.column_stats').alias('column_stats'),
                 col('stats.AVG').alias('AVG'),
                 col('stats.Q1').alias('Q1'),
                 col('stats.Q2').alias('Q2'),
                 col('stats.Q3').alias('Q3')
             )
```

```python
objectives = [
    struct( 
        col('team_1_won').alias('won'),
        col('team_1_towers_killed').alias('towers_killed'),
        col('team_1_first_tower').alias('first_tower'),
        col('team_1_atakhan').alias('atakhan'),
        col('team_1_baron_killed').alias('baron_killed'),
        col('team_1_dragon_killed').alias('dragon_killed'),
        col('team_1_grubs_killed').alias('grubs_killed'),
        col('team_1_herald').alias('herald')
    ),
    struct(
        col('team_2_won').alias('won'),
        col('team_2_towers_killed').alias('towers_killed'),
        col('team_2_first_tower').alias('first_tower'),
        col('team_2_atakhan').alias('atakhan'),
        col('team_2_baron_killed').alias('baron_killed'),
        col('team_2_dragon_killed').alias('dragon_killed'),
        col('team_2_grubs_killed').alias('grubs_killed'),
        col('team_2_herald').alias('herald')
    )
]

team_objectives_binned_df =  match_df_exploded.groupBy('game_id', 'game_duration')\
    .agg(
        first('team_1_won').alias('team_1_won'),
        first('team_1_towers_killed').alias('team_1_towers_killed'),
        first('team_1_first_tower').alias('team_1_first_tower'),
        first('team_1_atakhan').alias('team_1_atakhan'),
        first('team_1_baron_killed').alias('team_1_baron_killed'),
        first('team_1_dragon_killed').alias('team_1_dragon_killed'),
        first('team_1_grubs_killed').alias('team_1_grubs_killed'),
        first('team_1_herald').alias('team_1_herald'),
        first('team_2_won').alias('team_2_won'),
        first('team_2_towers_killed').alias('team_2_towers_killed'),
        first('team_2_first_tower').alias('team_2_first_tower'),
        first('team_2_atakhan').alias('team_2_atakhan'),
        first('team_2_baron_killed').alias('team_2_baron_killed'),
        first('team_2_dragon_killed').alias('team_2_dragon_killed'),
        first('team_2_grubs_killed').alias('team_2_grubs_killed'),
        first('team_2_herald').alias('team_2_herald')
    )\
    .withColumn('team_objectives', explode(array(*objectives)))\
    .select(
        col('game_id'),
        col('game_duration'),
        col('team_objectives.won').alias('won'),
        col('team_objectives.towers_killed').alias('towers_killed'),
        col('team_objectives.first_tower').alias('first_tower'),
        col('team_objectives.atakhan').alias('atakhan'),
        col('team_objectives.baron_killed').alias('baron_killed'),
        col('team_objectives.dragon_killed').alias('dragon_killed'),
        col('team_objectives.grubs_killed').alias('grubs_killed'),
        col('team_objectives.herald').alias('herald') 
    )\
    .withColumn('minutes_group', floor(((col('game_duration') / 60) / 5)) * 5)\
    .withColumn('minutes_group', concat(lit('['), col('minutes_group'), lit(' - '), col('minutes_group') + 5, lit(']')))


team_cols = [
    'towers_killed',
    'baron_killed',
    'dragon_killed',
    'grubs_killed',
    'herald'
]

agg_exprs = []
for col_name in team_cols :
    agg_exprs.append(round(avg(col_name), 4).alias(f"avg_{col_name}"))
    for quartile, label in zip([0.25, 0.5, 0.75], ["Q1", "Q2", "Q3"]):
        agg_exprs.append(round(percentile(col_name, lit(quartile), lit(1)), 4).alias(f"{label}_{col_name}"))
            
agg_team_data_df = team_objectives_binned_df\
    .groupBy('minutes_group', 'won')\
    .agg(
        count('*').alias('games_in_window'),
        sum(col('first_tower').cast(IntegerType())).alias('first_tower'),
        sum(col('atakhan').cast(IntegerType())).alias('atakhan'),
        *agg_exprs
    )\
    .withColumn('first_tower', round((col('first_tower') / col('games_in_window')) * 100, 2))\
    .withColumn('atakhan', round((col('atakhan') / col('games_in_window')) * 100, 2))\
    .orderBy(asc('minutes_group'), desc('won'))
```

```python
import builtins

correlated_cols = ['towers_killed', 'baron_killed', 'dragon_killed', 'grubs_killed', 'herald']
bool_cols = ['won', 'first_tower', 'atakhan']

for col_name in bool_cols:
    team_objectives_binned_df = team_objectives_binned_df.withColumn(col_name + "_num", col(col_name).cast("integer"))

all_num_cols = [c + "_num" for c in bool_cols] + correlated_cols

duration_groups = [r["minutes_group"] for r in team_objectives_binned_df.select("minutes_group").distinct().collect()]
duration_groups.sort()

by_group_corrs = {}

for g in duration_groups:
    subdf = team_objectives_binned_df.filter(col("minutes_group") == g)
    corrs = [(c, subdf.stat.corr("won_num", c)) for c in all_num_cols if c != "won_num"]
    by_group_corrs[g] = sorted(corrs, key=lambda x: builtins.abs(x[1]), reverse=True)
```

```python
team_objectvis_with_ff_df =  match_df_exploded.groupBy('game_id', 'game_duration', 'player_ended_surrender')\
    .agg(
        first('team_1_won').alias('team_1_won'),
        first('team_1_towers_killed').alias('team_1_towers_killed'),
        first('team_1_first_tower').alias('team_1_first_tower'),
        first('team_1_atakhan').alias('team_1_atakhan'),
        first('team_1_baron_killed').alias('team_1_baron_killed'),
        first('team_1_dragon_killed').alias('team_1_dragon_killed'),
        first('team_1_grubs_killed').alias('team_1_grubs_killed'),
        first('team_1_herald').alias('team_1_herald'),
        first('team_2_won').alias('team_2_won'),
        first('team_2_towers_killed').alias('team_2_towers_killed'),
        first('team_2_first_tower').alias('team_2_first_tower'),
        first('team_2_atakhan').alias('team_2_atakhan'),
        first('team_2_baron_killed').alias('team_2_baron_killed'),
        first('team_2_dragon_killed').alias('team_2_dragon_killed'),
        first('team_2_grubs_killed').alias('team_2_grubs_killed'),
        first('team_2_herald').alias('team_2_herald')
    )\
    .withColumn('team_objectives', explode(array(*objectives)))\
    .select(
        col('game_id'),
        col('game_duration'),
        col('player_ended_surrender').alias("ff"),
        col('team_objectives.won').alias('won'),
        col('team_objectives.towers_killed').alias('towers_killed'),
        col('team_objectives.first_tower').alias('first_tower'),
        col('team_objectives.atakhan').alias('atakhan'),
        col('team_objectives.baron_killed').alias('baron_killed'),
        col('team_objectives.dragon_killed').alias('dragon_killed'),
        col('team_objectives.grubs_killed').alias('grubs_killed'),
        col('team_objectives.herald').alias('herald')
    )\
    .withColumn('minutes_group', floor(((col('game_duration') / 60) / 5)) * 5)\
    .withColumn('minutes_group', concat(lit('['), col('minutes_group'), lit(' - '), col('minutes_group') + 5, lit(']')))

minute_window = Window.partitionBy('minutes_group')
minute_and_ff_window = Window.partitionBy('minutes_group', 'ff')

agg_exprs = []
for col_name in team_cols :
    agg_exprs.append(round(avg(col_name), 4).alias(f"avg_{col_name}"))
    for quartile, label in zip([0.25, 0.5, 0.75], ["Q1", "Q2", "Q3"]):
        agg_exprs.append(round(percentile(col_name, lit(quartile), lit(1)), 4).alias(f"{label}_{col_name}"))
            
agg_team_data_with_ff_df = team_objectvis_with_ff_df\
    .groupBy('minutes_group', 'won', 'ff')\
    .agg(
        count('*').alias('games_in_window'),
        sum(col('first_tower').cast(IntegerType())).alias('first_tower'),
        sum(col('atakhan').cast(IntegerType())).alias('atakhan'),
        *agg_exprs
    )\
    .withColumn('games_ff_in_window', (sum(col('games_in_window')).over(minute_and_ff_window) / 2))\
    .withColumn('games_in_window', (sum(col('games_in_window')).over(minute_window) / 2))\
    .withColumn('percents_ff_in_window', round((col('games_ff_in_window') / col('games_in_window')) * 100, 2))\
    .withColumn('first_tower', round((col('first_tower') / col('games_in_window')) * 100, 2))\
    .withColumn('atakhan', round((col('atakhan') / col('games_in_window')) * 100, 2))\
    .orderBy(asc('minutes_group'), desc('won'), desc('ff'))
```
